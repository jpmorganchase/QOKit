```python
from qokit import parameter_utils
from qokit.qaoa_circuit_labs import get_qaoa_circuit
from qokit.labs import get_gate_optimized_terms_greedy
```

# Generating QAOA circuits for LABS

Below we show how to generate (unoptimized) QAOA circuits for LABS in Qiskit. Further circuit optimizations would lead to better performance on hardware.

```python
N = 6
p = 1
gamma, beta = parameter_utils.get_best_known_parameters_for_LABS_wrt_overlap_for_p(N, p)
terms = get_gate_optimized_terms_greedy(N) # indices of Pauli Zs
```

```python
qc = get_qaoa_circuit(N, terms, gamma, beta, save_statevector=False)
```

```python
qc.draw()
```
